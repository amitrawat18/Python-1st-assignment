## **Question-1) Discuss String Slicing and provide examples.**

Answer-1)  String slicing in Python refers to the ability to extract a substring from a string by specifying a range of indices. It allows you to efficiently access and manipulate portions of a string.
The syntax for string slicing is:
  



```python
string[start:end:step]
```

start: The index of the first character to include in the slice (inclusive). If omitted, it defaults to 0.

end: The index of the character where the slice stops (exclusive). If omitted, it defaults to the end of the string.

step: The step size or stride. It determines the interval between characters to include in the slice. If omitted, it defaults to 1.

**Return Type**:-

The result of a slicing operation is always a string (str type) that contains a subset of the characters from the original string.

**String Slicing Examples**:-

Letâ€™s see how to use string slicing in Python with the examples below.

Retrieve All Characters
To retrieve the entire string, use slicing without specifying any parameters.

```python
s="PW SKILLS"
# now we will write the syntax to get the entire string
s[::]
```

**Explanation:**

 Using [::] without specifying start, end, or step returns the complete string.


 **Get All Characters Before or After a Specific Position:-**

To get all the items from a specific position to the end of the string, we can specify the start index and leave the end blank.
And to get all the items before a specific index, we can specify the end index while leaving start blank.

```python
a="Data Science"
# now to get the characters of the world science only we will write the following syntax

a[5:13]
# here the starting index is 5 and end index is 13 because python gives the value of a character less than the specified index
```

**Extract Characters Between Two Positions:-**

To extract characters between specific positions, provide both start and end indices.

```python
name="Ganesh"
#now if we want just the characters "nes" we will write
name[2:5]
```

**Get Characters at Specific Intervals**:-

To retrieve characters at regular intervals, use the step parameter.

```python
m= "qaretimovu"
# in the above word if we want to retrieve the vowels we will use
m[1:10:2]

# here 1 defines the start index, 10 defines the last index and 2 defines the intervals
```

##**Question-5)** **Describe the key features of sets and provide examples of their use.**

**Answer-5)**  In Python, a set is an unordered collection of unique elements. Sets are a fundamental data structure in Python, and they come with several key features that make them useful for various tasks.

Here are the key features of sets, along with examples to demonstrate their use:

## 1 Unordered Collection:-

* Sets do not maintain any order of elements. When you print a set, the order in which elements are displayed may not be the same as the order in which they were added.

```python
s={"Sanjay", "Ankit", "Rohit", 1, 34, True}
print(s)
#we can see that elements are arranged in unordered way
```

##2. Unique Elements:-
A set automatically removes duplicate elements. If you try to add a duplicate element to a set, it will not be added.

```python
sets={1,1,1,2,2,2,23,3, "Amit", "Ankit", "Amit"}
sets
```

## 3. Mutable:-
* Sets are mutable, meaning you can add or remove elements after the set is created

```python
set1={23, 45, "Dev", True, False}
# to remove a element w use.remove()
set1.remove(45)
# to add a element we use .add(90)
set1.add(90)
print(set1)
```

## 4. No Indexing, Slicing, or Looping with Indexes:-

* Sets do not support indexing, slicing, or accessing elements by position, because they are unordered

 Indexing is not allowed

print(set1[0])  # This would raise a TypeError: 'set' object is not subscriptable


## 5. Set Operations:-

Sets support mathematical set operations such as:

* Union (| or union())

* Intersection (& or intersection())

* Difference (- or difference())

* Symmetric Difference (^ or symmetric_difference())

```python
# Create two sets
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# Union (all unique elements from both sets)
union_set = set_a | set_b
print(union_set)  # Output: {1, 2, 3, 4, 5, 6}

# Intersection (elements common to both sets)
intersection_set = set_a & set_b
print(intersection_set)  # Output: {3, 4}

# Difference (elements in set_a but not in set_b)
difference_set = set_a - set_b
print(difference_set)  # Output: {1, 2}

# Symmetric Difference (elements in either set, but not both)
sym_diff_set = set_a ^ set_b
print(sym_diff_set)  # Output: {1, 2, 5, 6}
```

## 6. Set Comprehensions:-

Just like lists and dictionaries, sets support set comprehensions, which allow you to construct sets in a compact way.


```python
# Creating a set of squares using a set comprehension
r={2, 3, 4 ,5 ,6}
for square in r:
  print(square**2)
```

## 7. Membership Testing:-

Sets are optimized for membership testing, meaning checking if an element is present in a set or not.

```python
my_set = {1, 2, 3, 4}
print(5 in my_set)
print(2 in my_set)
```

## 8. Set Methods:-
Sets have a variety of built-in methods to perform operations like adding, removing, and clearing elements.

```python
my_set = {1, 2, 3}

# Add an element
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}

# Remove an element (raises KeyError if the element is not found)
my_set.remove(2)
print(my_set)  # Output: {1, 3, 4}

# Remove an element (does not raise an error if element is not found)
my_set.discard(5)  # No error even though 5 is not in the set
print(my_set)  # Output: {1, 3, 4}

# Pop an element (removes and returns an arbitrary element)
removed_element = my_set.pop()
print(removed_element)  # Output: 1 (or some other element, as sets are unordered)
print(my_set)  # Output: {3, 4}

# Clear the set (remove all elements)
my_set.clear()
print(my_set)  # Output: set()  (empty set)
```
